```python
import folium
from folium.plugins import MarkerCluster
from folium import Icon, Popup

# Suponiendo que tienes el DataFrame `spacex_df` con columnas: 'Lat', 'Long', 'class', 'Launch Site'

# 1. Crear una nueva columna 'marker_color' en el DataFrame `spacex_df` para determinar el color del marcador
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# 2. Crear el mapa base centrado en una ubicación general de los sitios de lanzamiento
site_map = folium.Map(location=[28.573255, -80.646895], zoom_start=5)

# 3. Crear el objeto MarkerCluster y añadirlo al mapa
marker_cluster = MarkerCluster().add_to(site_map)

# 4. Recorrer cada registro en `spacex_df` para crear un marcador
for index, record in spacex_df.iterrows():
    # Crear un marcador con el color basado en 'marker_color' y añadir un popup con el nombre y resultado del sitio de lanzamiento
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],  # Cambia 'Lat' y 'Long' por los nombres de tus columnas de coordenadas
        icon=Icon(color='white', icon_color=record['marker_color']),
        popup=Popup(f"Launch Site: {record['Launch Site']} - Result: {'Success' if record['class'] == 1 else 'Failure'}")
    )
    # Añadir el marcador al MarkerCluster
    marker_cluster.add_child(marker)

# 5. Mostrar el mapa
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Función para calcular la distancia entre dos coordenadas
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radio aproximado de la Tierra en km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Tomar las coordenadas del primer sitio de lanzamiento en el DataFrame `spacex_df`
launch_site_lat = spacex_df.iloc[0]['Lat']
launch_site_lon = spacex_df.iloc[0]['Long']

# Coordenadas de la costa más cercana (deberían ser definidas)
coastline_lat = 28.56367  # Ejemplo de latitud de la costa más cercana
coastline_lon = -80.57163  # Ejemplo de longitud de la costa más cercana

# Calcular la distancia entre el sitio de lanzamiento y la costa
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Crear el mapa centrado en el sitio de lanzamiento
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=15)

# Crear y añadir un marcador en el punto de la costa más cercano
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)
site_map.add_child(distance_marker)

# Dibujar una línea entre el sitio de lanzamiento y el punto de la costa
line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    weight=1,
    color='blue'
)
site_map.add_child(line)

# Mostrar el mapa
site_map
```

```python
import folium
from folium import DivIcon
from math import sin, cos, sqrt, atan2, radians
from folium.plugins import MarkerCluster

# Función para calcular la distancia entre dos coordenadas
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radio aproximado de la Tierra en km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordenadas de la costa más cercana (modifica manualmente según tus datos)
coast_point = (28.562278, -80.567856)  # Ejemplo de coordenadas de la costa

# Crear el mapa centrado en una ubicación promedio de los puntos de lanzamiento
site_map = folium.Map(location=[28.562302, -80.567709], zoom_start=15)

# Crear un objeto MarkerCluster para agrupar los marcadores
marker_cluster = MarkerCluster().add_to(site_map)

# Variable para rastrear el punto de lanzamiento más cercano
min_distance = float('inf')
closest_launch_coords = None

# Iterar sobre cada punto de lanzamiento en `spacex_df`
for index, row in spacex_df.iterrows():
    launch_site_lat = row['Lat']
    launch_site_lon = row['Long']
    launch_site_name = row['Launch Site']

    # Calcular la distancia entre el sitio de lanzamiento actual y el punto de la costa
    distance = calculate_distance(launch_site_lat, launch_site_lon, coast_point[0], coast_point[1])
    
    # Actualizar el sitio de lanzamiento más cercano a la costa
    if distance < min_distance:
        min_distance = distance
        closest_launch_coords = (launch_site_lat, launch_site_lon)

    # Crear y añadir un marcador al cluster de marcadores en el sitio de lanzamiento
    launch_marker = folium.Marker(
        location=[launch_site_lat, launch_site_lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 10px; color: {row["marker_color"]};"><b>{launch_site_name}</b></div>'
        )
    )
    marker_cluster.add_child(launch_marker)

# Crear y añadir un marcador en el punto de la costa más cercano
distance_marker = folium.Marker(
    location=coast_point,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(min_distance),
    )
)
site_map.add_child(distance_marker)

# Dibujar una línea entre el sitio de lanzamiento más cercano y el punto de la costa
if closest_launch_coords:
    line = folium.PolyLine(
        locations=[closest_launch_coords, coast_point],
        weight=2,
        color='blue'
    )
    site_map.add_child(line)

# Mostrar el mapa
site_map
```
